```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)
print(df.head())
```

```python
print("Forma del DataFrame:", df.shape)

print("\nPrimeras filas del DataFrame:")
print(df.head())

print("\nÚltimas filas del DataFrame:")
print(df.tail())

print("\nTipos de datos de las columnas:")
print(df.dtypes)

print("\nTipo del objeto DataFrame:", type(df))
```

```python
df = pd.read_csv(url)

description = df.describe()

print(description)
```

```python
min_values = df.min()
max_values = df.max()

# Mostrar los valores mínimos y máximos
print("Valores mínimos:")
print(min_values)
print("\nValores máximos:")
print(max_values)
```

```python

df['Date'] = pd.to_datetime(df['date'])


df_sorted = df.sort_values(by='date')

print(df_sorted.head())
```

```python
Garfield_Dan_Ryan = df[df['stationname'] == 'Garfield-Dan Ryan']
print(Garfield_Dan_Ryan.head())
```

```python
Garfield_South_Elevated = df[df['stationname'] == 'Garfield-South Elevated']

print(Garfield_South_Elevated.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(rc={'figure.figsize':(11, 4)})

plt.figure(figsize=(11, 4))
plt.plot(Garfield_Dan_Ryan.index, Garfield_Dan_Ryan['rides'], linewidth=0.5)
plt.title('Número de Viajes en la estación Garfield-Dan Ryan')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Viajes')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(rc={'figure.figsize':(11, 4)})

plt.figure(figsize=(11, 4))
plt.plot(Garfield_South_Elevated.index, Garfield_South_Elevated['rides'], linewidth=0.5)
plt.title('Número de Viajes en la estación Garfield-South Elevated')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Viajes')
plt.grid(True)
plt.show()
```
